```python
folder_path = "../data"
file_path = "test.txt"

full_path = os.path.join(folder_path, file_path)
sample_texts = {
    full_path: """This is a sample text file. It contains multiple lines of text.
 The purpose of this file is to demonstrate how to read and write files in Python. 
 You can add more lines to this file as needed."""
}

for path, content in sample_texts.items():
    with open(path, "w") as f:
        f.write(content)
```

```python
with open(full_path, "r") as f:
    content = f.read()
print(content)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader(full_path)
document = loader.load()
document
```

```python
from langchain_community.document_loaders import PyPDFLoader
import os
folder_path = "../data/pdf"
file_path = "pdf_10.pdf"
full_path = os.path.join(folder_path, file_path)
pdf_loader = PyPDFLoader(full_path)
pdf_document = pdf_loader.load()
pdf_document[0].page_content
```

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain_community.document_loaders import TextLoader
folder_path = "../data/pdf"
loader = DirectoryLoader(
    path = folder_path,
    glob = "*.pdf",
    loader_cls = PyMuPDFLoader,
)

documents = loader.load()
documents[0].metadata
```
